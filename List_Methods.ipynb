List Methods

```python
fruits = ["apple","banana","cherry","orange"]
```

```python
type(fruits)
```

```python
print(fruits)
```

.append : Adds an element to the end of the list

```python
fruits.append("greeps")
```

```python
fruits
```

.insert : Inserts an element at the specified index

```python
fruits.insert(2,"greeps")
```

```python
fruits
```

.Extend : adds all elements from an iterable to the end of the list


```python
fruits = ["orange","apple"]
more_fruits = ["greeps","cherry","banana"]
```

```python
fruits.extend(more_fruits)
```

```python
fruits
```

.remove : Removes the first occurrence of the specified element from the list

```python
fruits = ["apple","banana","cherry","orange"]
```

```python
fruits.remove("banana")
```

```python
fruits
```

.pop : removes and returns the elements at the specified index. if no index is provied, it removes and returns the last element

```python
fruits = ["apple","banana","cherry","orange"]
removed_fruits = fruits.pop(2)
```

```python
fruits
```

```python
removed_fruits
```

.clear : Remove all elements from the list

```python
fruits = ["apple","banana","cherry","orange"]
```

```python
fruits.clear()
```

```python
fruits
```

.reverse : Reverse the order of the elements in the list

```python
fruits = ["apple","banana","cherry","orange"]
```

```python
fruits = ["apple","banana","cherry","orange"]
```

```python
fruits_copy = fruits.copy()
```
