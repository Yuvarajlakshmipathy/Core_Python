```python
                                                             # Compound Intrest
p = int (input("Enter principle Amount : "))
n = int (input("Enter Number of Year's : "))
r = float (input ("Enter Rate of Intrest : "))
Full_Amount = p*(1+(r/100))**n
print("Full Amount :",Full_Amount)
Intrest = Full_Amount - p
print("Intrest : ",Intrest)
```

```python
                                                              #Factorial Number
Num = int(input ("Enter Any Number :"))
Factorial = 1
if Num < 0:
  print(Num,"Factorial Not Found")
elif Num == 0:
  print("Factorial is 1")
else:
  for i in range (1,Num + 1):
    Factorial = Factorial * i
    print(i, "Factorial", Factorial)
```

```python
                                                          # Prime Number Finding
Number = int(input("Enter Any Number : "))
if Number == 1:
    print(Number,"is Nutral Number !")
elif Number > 0:
  for i in range(2,Number):
    if (Number%i) == 0:
      print(Number,"is not a prime number !")
      break
  else:
      print(Number,"is prime number !")

else:
    print(Number,"is not a prime number !")
```

```python
                                                                # Perfect Number
Num = int (input("Enter any number : "))         # 6,28,496,8128 .... perfect number
sum = 0
for i in range (1,Num):
  if (Num % i) == 0:
    sum = sum + i
if (sum == Num):
  print(Num, "is a perfact number !")
else:
  print(Num, "is not perfact number !")
```

```python
                                                             # palindrome number
str = input("Enter your string: ")
length = len(str)
str = str.lower()
mid = length//2
rev = -1
for i in range(mid) :
    if str[i] == str[rev]:
        i +=1
        rev -=1
    else :
        print (str, "is not a palindrome")
        break
else:
    print(str, "is a palindrome")
```
