.get : Retrives the value associated with a given kry. if tthe key is not found, it returns the specified default value.

```python
my_dict = {"name":"yuvaraj","age":"24","gender":"male"}
```

```python
name = my_dict.get("name","unknown")
```

```python
name
```

```python
email = my_dict.get("email","not specified")
```

```python
email
```

.clear : Removes all the keys - value pairs from the dictionary, making it empty

```python
my_dict = {"name":"yuvaraj","age":"24","gender":"male"}
```

.pop : Removes and returns the value associated with the given key. if the key is not found , it returns the specified default value or raises a key error

```python
my_dict = {"name":"yuvaraj","age":"24","gender":"male"}
```

```python
my_dict.pop("gender")
```

```python
my_dict.pop("email","Not Specified")
```

```python
my_dict = {"name":"yuvaraj","age":"24","gender":"male"}
```

```python
copy_dict = my_dict.copy()
```

```python
copy_dict
```

```python
my_dict
```

.popitem : Removes and returns a key-value pair as a tuple in LIFO (Last in first out) order. Raused a keyError if the dictionary is empty.

```python
my_dict = {"name":"yuvaraj","age":"24","gender":"male"}
```

```python
item = my_dict.popitem()
```

```python
item
```

```python
my_dict
```

.setdefault : Returns the value associated with the given key. If the key is not found , it inserts the key-value pair into the dictionary with the specified default value

```python
my_dict = {"name":"yuvaraj","age":"24","gender":"male"}
```

```python
name = my_dict.setdefault("name","unknown")
```

```python
name
```

```python
email = my_dict.setdefault("email","Not Specified")
```

```python
email
```
